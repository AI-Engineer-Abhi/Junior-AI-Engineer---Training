Q16. Encapsulation — Bank Account
Create a class BankAccount where the balance is kept private. Create methods deposit(), withdraw(),
and check_balance().

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   # private attribute

    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited ₹{amount}")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn ₹{amount}")
        else:
            print("Insufficient balance")

    def check_balance(self):
        print(f"Current Balance: ₹{self.__balance}")


account = BankAccount(1000)

account.check_balance()
account.deposit(500)
account.withdraw(300)
account.check_balance()
```

Q19. Abstraction — Model
Create an abstract class Model with an abstract method train(). Create two child classes that
implement it.

```python
from abc import ABC, abstractmethod


class Model(ABC):

    @abstractmethod
    def train(self):
        pass


class SpamDetector(Model):
    def train(self):
        print("Training Spam Detection Model")


class FaceRecognition(Model):
    def train(self):
        print("Training Face Recognition Model")


spam = SpamDetector()
face = FaceRecognition()

spam.train()
face.train()
```

Q20. Hiding Data (Conceptual)
Which OOP pillar is used when we hide data from users? Explain with one real-world example. (No
coding required.)

The OOP pillar used to hide data from users is Encapsulation.

Encapsulation is the process of wrapping data and methods together in a class and restricting direct access to certain data. This helps protect the data from unauthorized modification and ensures it is accessed only through controlled methods.

Real-World Example: Bank Account

Consider a bank account.

The account balance is not directly accessible to customers.
Customers cannot modify the balance by typing a new value.
Instead, they use operations such as:
Deposit money
Withdraw money
Check balance

The bank's system controls how the balance is updated and validates each transaction.

This is an example of encapsulation, where the data (balance) is hidden and can only be accessed or modified through specific methods.

the same accuracy.

```python
class AIModel:
    def __init__(self, model_name, accuracy):
        self.model_name = model_name
        self.accuracy = accuracy

    def __eq__(self, other):
        return self.accuracy == other.accuracy


model1 = AIModel("Spam Detector", 95)
model2 = AIModel("Face Recognition", 95)

print(model1 == model2)
```

Q29. AI Login System Logging
Create an AI Login System. Whenever a user logs in, store the Time, Username, and Status inside
system.log using the logging module.

```python
import logging

# Configure logging
logging.basicConfig(
    filename="system.log",
    level=logging.INFO,
    format="%(asctime)s - %(message)s"
)

username = input("Enter username: ")
status = input("Enter login status (Success/Failed): ")

logging.info(f"Username: {username}, Status: {status}")

print("Login information saved to system.log")

# import logging → Imports the logging module.
# logging.basicConfig() → Configures where and how logs are stored.
# filename → Specifies the log file.
# logging.info() → Writes an informational log entry.
```
