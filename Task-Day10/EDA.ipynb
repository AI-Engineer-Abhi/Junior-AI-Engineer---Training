```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Loading the data
data = pd.read_csv("/content/data.csv")
```

```python
# Converting to DataFrame
df = pd.DataFrame(data)
```

Most of the columns are of object (categorical) data type, indicating that the dataset mainly consists of categorical information such as players, clubs, competitions, and venues.

```python
# Missing values in each column
df.isnull().sum()
```

```python
# Find any duplicates
print("Duplicates:", df.duplicated().sum())
```

No duplicate records were found in the dataset.

```python
# Univariate --- > Player Distribution
df["Player"].value_counts().plot(kind="bar", figsize=(8,5))
plt.title("Number of Goals by Player")
plt.xlabel("Player")
plt.ylabel("Count")
plt.show()
```

```python
# Goals across seasons
# (How many Goals were scored throughtout the seasons?)
df["Season"].value_counts().plot(kind="line", figsize=(8,5))
plt.title("Goals across seasons")
plt.xlabel("Season")
plt.ylabel("Count")
```

The line plot shows how the number of goals varies across different seasons, helping identify periods with relatively higher or lower scoring activity.

```python
# Competition Distribution
df["Competition"].value_counts().plot(kind="bar", figsize=(8,5))
plt.title("Goals by Competition")
plt.xlabel("Competition")
plt.ylabel("Count")
plt.show()
```

```python
# Venue Distribution
df["Venue"].value_counts().plot(kind="pie", autopct="%1.1f%%")
# plt.ylabel("")
plt.title("Home vs Away Goals")
plt.show()
```

The pie chart illustrates the proportion of goals scored at Home and Away venues, showing whether one venue has a higher frequency.

```python
# Club Distribution
df["Club"].value_counts().plot(kind = "bar", figsize=(8,5))
plt.title("Club Distribution")
plt.xlabel("Club")
plt.ylabel("Number of Goals")
plt.show()
```

```python
# Bivariate Analysis --> Player VS Competition
#  (Which player scored in which competition?)
player_competition = df.groupby(["Player", "Competition"]).size()

player_competition.plot(kind="bar", figsize=(10,5))

plt.title("Player vs Competition")
plt.xlabel("Player")
plt.ylabel("Number of Goals")
plt.xticks(rotation=45)
plt.show()
```

```python
# Venue VS Competition
venue_competition = df.groupby(["Venue", "Competition"]).size()

venue_competition.plot(kind="bar", figsize=(8,5))

plt.title("Venue vs Competition")
plt.xlabel("Venue")
plt.ylabel("Count")
plt.show()
```

The comparison shows how goals are distributed between Home and Away venues across different competitions.

```python
# Club VS Player
club_player = df.groupby(["Club", "Player"]).size()

club_player.plot(kind="bar", figsize=(8,5))

plt.title("Club vs Player")
plt.xlabel("Club")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Club VS Competition (Which clubs scored in which competitions?)
club_competition = df.groupby(["Club", "Competition"]).size()

club_competition.plot(kind="bar", figsize=(8,5))

plt.title("Club vs Competition")
plt.xlabel("Club")
plt.ylabel("Count")
plt.show()
```

Clubs have participated in multiple competitions, with some clubs contributing more goal records than others.

```python
# Multivariate Analysis  ---> Player VS Competition VS Venue
# (Which player scored in which competition at Home and Away?)

player_competition_venue = df.groupby(
    ["Player", "Competition", "Venue"]
).size().unstack()

player_competition_venue.plot(kind="bar", figsize=(10,5))

plt.title("Player vs Competition vs Venue")
plt.xlabel("Player and Competition")
plt.ylabel("Number of Goals")
plt.xticks(rotation=45)
plt.show()
```

```python
club_competition_venue = df.groupby(
    ["Club", "Competition", "Venue"]
).size().unstack(fill_value=0)

print(club_competition_venue)
```

```python
club_competition_venue.plot(
    kind="bar",
    figsize=(10,5),
    stacked=True
)

plt.title("Club vs Competition vs Venue")
plt.xlabel("Club and Competition")
plt.ylabel("Number of Goals")
plt.xticks(rotation=45)
plt.show()
```
