**Section A — Theoretical Questions**

 What is Natural Language Processing (NLP)? Explain in your own words how it acts
as a bridge between human language and computers.

Natural Language Processing (NLP) is a branch of Artificial Intelligence that enables computers to understand, interpret, and generate human language.

It acts as a bridge between humans and computers by converting natural language, such as English or any other spoken language, into a format that computers can understand and process. This allows us to interact with computers using normal language instead of writing complex programming instructions.
ex: chatbots, voice assistants

 Why do computers need NLP to understand human language? Give one real-world
example to support your answer.

Computers process information in the form of numbers and binary data, whereas humans communicate using words, sentences, emotions, and context. Because of this difference, computers cannot naturally understand human language. NLP helps convert text or speech into structured data that machines can analyze and interpret.

A good real-world example is an email spam filter. NLP analyzes the content of emails and identifies words or patterns such as "Congratulations," "Win now," or "Claim your prize." Based on these patterns, it classifies the email as spam or genuine.

 List any four real-life applications of NLP and briefly describe what each one does.

Chatbots and Virtual Assistants – Applications like ChatGPT and Siri understand user queries and provide relevant responses in natural language.

Machine Translation – Services such as Google Translate convert text from one language to another while preserving the meaning.

Spam Email Detection – NLP examines the content of emails to identify spam messages and prevent them from reaching the inbox.

Sentiment Analysis – Businesses analyze customer reviews and social media posts to determine whether opinions are positive, negative, or neutral.

What is a rule-based NLP system? Explain using the spam detection example
(checking if "Congratulations" is present in a message).

A rule-based NLP system works by following predefined rules instead of learning from data. It checks whether specific conditions are satisfied and then performs an action.

For example, suppose we create a rule that says:

If the message contains the word "Congratulations", mark it as Possible Spam.
Otherwise, mark it as Not Spam.

If the message is "Congratulations! You have won a free gift.", the system finds the keyword "Congratulations" and classifies it as spam. This approach is simple and easy to implement but may fail if the spam message uses different words that are not included in the rules.

 List and briefly explain the seven steps of the NLP pipeline, in the correct order.

Collect Text Data – Gather text from sources such as documents, websites, emails, or social media.

Text Cleaning – Remove unwanted characters, punctuation, HTML tags, emojis, and special symbols to make the text cleaner.

Convert to Lowercase – Convert all text to lowercase so that words like "Phone" and "phone" are treated as the same word.

Tokenization – Split the text into smaller units called tokens, usually words or sentences.

Remove Stopwords – Remove commonly used words such as "the," "is," and "and" because they usually add little meaning.

Feature Extraction – Convert text into numerical features using methods like Bag of Words or TF-IDF so that machine learning models can process it.

Model Building and Prediction – Train a machine learning model on the processed data and use it to make predictions such as spam detection or sentiment classification.

 What is tokenization? Why is it a necessary step before a machine learning model
can process text?

Tokenization is the process of splitting text into smaller units called tokens. These tokens are usually individual words but can also be sentences or characters.

For example:

Sentence: "I love Artificial Intelligence."

Tokens: ["I", "love", "Artificial", "Intelligence"]

Tokenization is necessary because machine learning models cannot understand complete sentences directly. Breaking text into tokens makes it easier to analyze, count word frequencies, and convert text into numerical features for model training.

 What are stopwords? Give three examples of stopwords in English and explain why
they are usually removed.

Stopwords are common words that appear frequently in a language but usually do not add significant meaning to the text during analysis.

Examples of English stopwords are:

the
is
and

These words are usually removed because they occur very often and do not help distinguish one document from another. Removing them reduces unnecessary data, improves processing speed, and allows the model to focus on more meaningful words.

Why is it important to convert all text data to lowercase before processing it in NLP?
Illustrate with an example.

Converting text to lowercase ensures that words with different capitalizations are treated as the same word. This reduces duplicate entries and improves the accuracy of text analysis.

For example, consider the sentence:

"Phone phone PHONE"

Without converting to lowercase, the computer treats them as three different words:

Phone
phone
PHONE

After converting to lowercase:

phone
phone
phone

Explain the difference between using = and == while filtering rows in a DataFrame,
and why mixing them up causes errors.

The operator = is used for assignment, which means it assigns a value to a variable. The operator == is used for comparison, which checks whether two values are equal.

For example, to filter positive reviews:

df[df["Sentiment"] == ["Positive"]
Here, == checks each row to see whether the value in the Sentiment column is "Positive".

If we mistakenly write:
df[df["Sentiment"] = ["Positive"]
Python produces a syntax error because = cannot be used for comparison inside a filtering condition. Mixing these two operators is a common beginner mistake and prevents the program from running correctly.
