Q14. Convert image to HSV and display H, S, V channels separately

```python
import cv2
import matplotlib.pyplot as plt

# Read image
image = cv2.imread(image_filename)

# Convert BGR to HSV
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Split channels
H, S, V = cv2.split(hsv)                           #to separate these three channels into individual grayscale images for visualization.

# Display channels
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(H, cmap="gray")
plt.title("Hue")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(S, cmap="gray")
plt.title("Saturation")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(V, cmap="gray")
plt.title("Value")
plt.axis("off")

plt.show()

# HSV stands for

# H (Hue) → the actual color (red, green, blue, etc.)
# S (Saturation) → how pure or vivid the color is
# V (Value) → the brightness of the color

# For example:

# Hue identifies which color you are seeing.
# Saturation tells you whether that color is vivid or faded.
# Value tells you how bright or dark it appears.
```

Q16. Apply Gaussian Blur, Median Blur, and Bilateral Filter

```python
import cv2
import matplotlib.pyplot as plt

# Read image
image = cv2.imread(image_filename)

# Convert to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Apply filters
gaussian = cv2.GaussianBlur(image, (5, 5), 0)        # (5, 5) --> Kernel size
                                                     # 0 --> sigma controls how much weight nearby pixels receive
                                                    #  OpenCV automatically calculates the best sigma from the kernel size, if you set it to 0.
median = cv2.medianBlur(image, 5)


bilateral = cv2.bilateralFilter(image, 9, 75, 75)     # 9 --> diameter
                                                      # sigma color -->how different two colors can be while still being averaged together.
                                                      # small sigma --> Only nearly identical colors are mixed. Edges remain sharp.



# Display images
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(gaussian)
plt.title("Gaussian Blur")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(median)
plt.title("Median Blur")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(bilateral)
plt.title("Bilateral Filter")
plt.axis("off")

plt.show()
```

Q17.
Write code to threshold a grayscale image using both THRESH_BINARY and
THRESH_BINARY_INV, and display both outputs.

```python
import cv2
import matplotlib.pyplot as plt

# Read image
image = cv2.imread(image_filename)

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply threshold
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
                                                                       # If pixel > 127, it becomes 255, otherwise it becomes 0.
_, binary_inv = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)

# Display images
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)               #subplot(rows, columns, position)
plt.imshow(binary, cmap="gray")
plt.title("Binary")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(binary_inv, cmap="gray")
plt.title("Binary Inverse")
plt.axis("off")

plt.show()
```

Q21.
Write Keras code to build a CNN with two Conv2D + MaxPooling blocks, a Flatten
layer, and two Dense layers ending in a Softmax output.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.layers import Flatten, Dense

# Create model
model = Sequential()         # Sequential() creates a neural network where layers are added one after another.

# First Convolution Block
model.add(Conv2D(
    filters=32,            # Create 32 different feature maps.
    kernel_size=(3,3),     # window size
    activation="relu",
    input_shape=(224,224,3)  # RGB
))

model.add(MaxPooling2D(pool_size=(2,2)))

# Second Convolution Block
model.add(Conv2D(
    filters=64,
    kernel_size=(3,3),
    activation="relu"
))

model.add(MaxPooling2D(pool_size=(2,2)))

# Convert feature maps into a vector
model.add(Flatten())

# Hidden Layer
model.add(Dense(128, activation="relu"))

# Output Layer
model.add(Dense(10, activation="softmax"))

# Display model
model.summary()
```

Write code to compile a CNN model using the Adam optimizer with a learning rate of

```python
from tensorflow.keras.optimizers import Adam

model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)
```
