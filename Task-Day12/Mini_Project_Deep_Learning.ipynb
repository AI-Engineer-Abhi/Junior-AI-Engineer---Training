**Task 5 — Write the forward pass function**

Write a function forward(X, weights, biases) that: (1) computes the weighted sum for the hidden
layer, (2) applies ReLU, (3) computes the weighted sum for the output layer, and (4) applies the linear
activation. Run it once on the first 4 rows of your dataset and print the predicted scores — they will look
random right now, since the network hasn't learned anything yet.

```python
# Forward Propagation

def forward_propagation(x, w1, b1, w2, b2):

  # Hidden layer weighted sum
  z1 = np.dot(x, w1) + b1

  # Hidden layer Activation
  a1 = relu(z1)

  # Output layer weighted sum
  z2 = np.dot(a1, w2) + b2

  # Output layer activation
  a2 = linear(z2)

  return z1, a1, z2, a2

# Performs one forward pass through the neural network.
#
# Parameters:
#     X  : Input data
#     W1 : Input to Hidden weights
#     b1 : Hidden layer biases
#     W2 : Hidden to Output weights
#     b2 : Output layer bias
#
# Returns:
#     Z1 : Hidden layer weighted sum
#     A1 : Hidden layer activation (ReLU)
#     Z2 : Output layer weighted sum
#     A2 : Final prediction
```

```python
# Testing the function

# Test input
x_test = np.array([[10]])

# Forward pass
z1, a1, z2, a2 = forward_propagation(x_test, w1, b1, w2, b2)

print("Hidden weighted sum: ", z1)
print("After relu activation: ", a1)
print("Output layer weighted sum: ", z2)
print("After Linear Activation: ", a2)
```

**Task 6 — Implement the loss function**

Write a function mse_loss(y_true, y_pred) that returns the Mean Squared Error: the average of
(y_true - y_pred)². Run it on your Task 5 predictions versus the real exam scores and print the loss value.

```python
# Mean squared error loss

def mse_loss(y_true, y_pred):
  return np.mean((y_true - y_pred) ** 2)

# Calculates the Mean Squared Error (MSE).
#
#   Parameters:
#       y_true : Actual target values
#       y_pred : Predicted values
#
#   Returns:
#         Mean Squared Error
```

```python
# Testing the function

# Actual exam score
y_true = np.array(([80], [70], [90], [60]))

# Predicted score
y_pred = np.array(([75], [73], [88], [55]))

loss = mse_loss(y_true, y_pred)

print("MSE Loss: ", loss)
```

output layers), using the chain rule through the ReLU and linear activations. Print the shape of each

```python
# Backpropagation

# Performs one step of backpropagation and updates the weights.

#     Parameters:
#         X, Y : Input data and actual output
#         Z1, A1 : Hidden layer values
#         Z2, A2 : Output layer values
#         W1, b1 : Input -> Hidden parameters
#         W2, b2 : Hidden -> Output parameters
#         learning_rate : Step size

#     Returns:
#         Updated W1, b1, W2, b2


def back_propagation(x, y, z1, a1, z2, a2, w1, w2, b1, b2, learning_rate = 0.01):

  m = x.shape([0])  # Number of training examples

  # Output layer gradients
  dz2 = a2 - y
  dw2 = np.dot(a1.T, dz2) / m
  db2 = np.sum(dz2, axis = 0, keepdims = True)/m

  # Gradient descent update
  W2 = W2 - learning_rate * dW2
  b2 = b2 - learning_rate * db2

  W1 = W1 - learning_rate * dW1
  b1 = b1 - learning_rate * db1

  return W1, b1, W2, b2
```
