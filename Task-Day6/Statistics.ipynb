Part G – Probability & Business Analytics (35–40)

Q35. Calculate the probability that an employee has left the company.

```python
employees_left = (df["Left_Company"] == "Yes").sum()

total_employees = len(df)         # Total number of rows

probability = employees_left / total_employees

print("Probability that an employee has left the company:", probability)
```

Q36. Calculate the probability that an employee belongs to the AI department.

```python
ai_employees = (df["Department"] == "AI").sum()

total_employees = len(df)

probability = ai_employees / total_employees

print("Probability that an employee belongs to the AI department:", probability)
```

Q37. Calculate the conditional probability: P(Left Company | Performance Rating £
3).

```python
probability = (
    ((df["Performance_Rating"] <= 3) & (df["Left_Company"] == "Yes")).sum()
    / (df["Performance_Rating"] <= 3).sum()
)

print("Conditional Probability:", probability)
```
